```python
# local library
sys.path.append('../../../force_fields/statmechlib')
from statmechlib.read_write import read_lattice_ising
```

```python
sim_dir = '../src/mcsim/mini'
temps = ['t1', 't2', 't5', 't10', 't20', 't100']  # ferromagnetic
temps.extend(['a1', 'a2', 'a5', 'a10', 'a20', 'a100'])  # antiferromagnetic
```

```python
trjs = {}
for t in temps:
    print(t)
    trjs[t] = read_lattice_ising(os.path.join(sim_dir, t))

trjs.keys()
```

```python
cfgs = {}
cfgs_pad = {}
for key, trj in trjs.items():
    cfgs[key] = []
    cfgs_pad[key] = []
    for t_cfg, xyz_cfg in zip(trj['atom_type'], trj['xyz_latt']): 
        cfg = np.empty((15, 15), dtype=int)
        for t, (i, j, k) in zip(t_cfg, xyz_cfg):
            cfg[i-1, j-1] = t
        cfgs[key].append(cfg)
        cfgs_pad[key].append(np.pad(cfg, pad_width=((1,1), (1,1)), mode='wrap'))
```

```python
fig, axs = plt.subplots(nrows=2, ncols=6, figsize=(25,10))
axs = axs.flatten()
axs[0].matshow(cfgs['t1'][1000])
axs[1].matshow(cfgs['t2'][1000])
axs[2].matshow(cfgs['t5'][1000])
axs[3].matshow(cfgs['t10'][1000])
axs[4].matshow(cfgs['t20'][1000])
axs[5].matshow(cfgs['t100'][1000])
axs[6].matshow(cfgs['a1'][1000])
axs[7].matshow(cfgs['a2'][1000])
axs[8].matshow(cfgs['a5'][1000])
axs[9].matshow(cfgs['a10'][1000])
axs[10].matshow(cfgs['a20'][1000])
axs[11].matshow(cfgs['a100'][1000])
fig.tight_layout()
```
